```python
import sqlite3
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

csv_file_path = '/Users/jessiefreelander/Flatiron_labs/Git_lab_Phase Project/Phase-2-Project/Data/tn.movie_budgets.csv'

data = pd.read_csv(csv_file_path, sep=',')

data.head(5)
```

```python
print(data.isna().any())
```

```python
data = data.rename(columns = { 'release_date':'Release Date', 'movie':'Movie', 'production_budget' : 'Production Budget', 'domestic_gross': 'Domestic Gross', 'worldwide_gross':'World Wide Gross'})

data.head(1)
```

```python
data['Production Budget'] = data['Production Budget'].str.replace('$','')
data['Production Budget'] = data['Production Budget'].str.replace(',','')
data['Production Budget'] = data['Production Budget'].astype(int)
data.head(1)
```

```python
data['Domestic Gross'] = data['Domestic Gross'].str.replace('$','')
data['Domestic Gross'] = data['Domestic Gross'].str.replace(',','')
data['Domestic Gross'] = data['Domestic Gross'].astype(int)
data.head(1)
```

```python
data['World Wide Gross'] = data['World Wide Gross'].str.replace('$', '')
data['World Wide Gross'] = data['World Wide Gross'].str.replace(',', '')
data['World Wide Gross'] = data['World Wide Gross'].astype(int)
data.head(1)
```

```python
(data==0).any()
```

```python
Wzero_drop = data[data['World Wide Gross'] == 0].index
data = data.drop(index = Wzero_drop)
data = data.reset_index(drop=True)
Dzero_drop = data[data['Domestic Gross'] == 0].index
data = data.drop(index = Dzero_drop)
data = data.reset_index(drop=True)
data.head()
```

```python
(data==0).any()
```

```python
data['Domestic Profit Ratio'] = data['Domestic Gross'] / data['Production Budget']
data['World Wide Profit Ratio'] = data['World Wide Gross'] / data['Production Budget']
data.head(1)
```

```python
query = ("""
SELECT p.primary_name AS Director, 
       mb.genres AS Genres, 
       mb.start_year AS Release_Year, 
       mb.original_title AS Movie,
       mr.averagerating AS Rating
FROM directors AS d
JOIN movie_basics AS mb ON d.movie_id = mb.movie_id
JOIN persons AS p ON d.person_id = p.person_id
JOIN movie_ratings AS mr ON mb.movie_id = mr.movie_id;
""")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Filter the dataset to include only the top 10 directors
top_10_directors = exploded_data[exploded_data['Director'].isin(top_10_Mystery_directors['Director'])]

# Step 2: Calculate the mean and confidence interval for Domestic Profit Ratio
directors_stats = top_10_directors.groupby('Director').agg({
    'Domestic Profit Ratio': ['mean', 'count', 'std']  # Mean, count, and standard deviation
}).reset_index()

# Flatten multi-level columns
directors_stats.columns = ['Director', 'Mean Domestic Profit Ratio', 'Count', 'STD']

# Step 3: Calculate the standard error of the mean (SEM)
directors_stats['SEM'] = directors_stats['STD'] / np.sqrt(directors_stats['Count'])

# Step 4: Calculate the 95% confidence interval (CI)
confidence_level = 1.96  # For 95% confidence level
directors_stats['CI Lower'] = directors_stats['Mean Domestic Profit Ratio'] - confidence_level * directors_stats['SEM']
directors_stats['CI Upper'] = directors_stats['Mean Domestic Profit Ratio'] + confidence_level * directors_stats['SEM']

# Step 5: Sort directors by Mean Domestic Profit Ratio in descending order
directors_stats = directors_stats.sort_values(by='Mean Domestic Profit Ratio', ascending=False)

# Step 6: Plot the bar chart with error bars (confidence intervals)
plt.figure(figsize=(12, 6))

# Plot the mean domestic profit ratio with error bars (confidence intervals)
plt.bar(directors_stats['Director'], directors_stats['Mean Domestic Profit Ratio'], yerr=[directors_stats['Mean Domestic Profit Ratio'] - directors_stats['CI Lower'], 
                                                    directors_stats['CI Upper'] - directors_stats['Mean Domestic Profit Ratio']],
        capsize=5, color='skyblue')

# Add labels and title
plt.xlabel('Director')
plt.ylabel('Mean Domestic Profit Ratio')
plt.title('Top 10 Mystery Directors by Mean Domestic Profit Ratio with Confidence Intervals')
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```
